```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("bank.csv")
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.histplot(df['balance'], bins=30, kde=True)
plt.title("Balance Distribution")
plt.show()
```

```python
job_counts = df['job'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(job_counts, labels=job_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Job Distribution')
plt.show()
```

```python
df['balance'].plot(kind='hist',bins=30,color='blue')
plt.title("Balance Distribution")
plt.show()
```

```python
df['age'].plot(kind='hist',bins=20,color='green')
plt.title("Age Distribution")
plt.show()
```

```python
df['marital'].value_counts().plot(kind='bar',color='orange')
plt.title("Marital Status Distribution")
plt.show()
```

```python
df['job'].value_counts().plot(kind='bar',color='purple')
plt.title("Job Distribution")
plt.show()
```
